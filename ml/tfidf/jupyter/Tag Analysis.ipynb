```python
import pandas as pd
import pymysql

pd.set_option('display.max_rows', 1000)
```

```python
HOST = 'yew-review-test.c2usv7nad3n6.us-west-2.rds.amazonaws.com'
USER = 'admin'
PASSWORD = 'Y3wR3view9!'
DATABASE = 'yewreview'

query = """
  SELECT at.tagId,
         t.name, 
         AVG(weight) as meanWeight,
         COUNT(*) as articleWithTagCount
    FROM ArticleTag at
    JOIN Tag t ON at.tagId = t.id
   GROUP BY at.tagId, t.name
"""

try:
    connection = pymysql.connect(HOST, USER, PASSWORD, DATABASE)
    df = pd.read_sql(query, connection)
finally:
    if connection:
        connection.close()
```

```python
(
    df
#     .sort_values('articleWithTagCount', ascending=False)
    .set_index('tagId')
    .assign(tokenCount=lambda df: df.name.str.split(' ').str.len())
    .sort_values(['tokenCount', 'articleWithTagCount'], ascending=False)
    .iloc[:100]
)
```
